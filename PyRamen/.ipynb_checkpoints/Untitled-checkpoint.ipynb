```python
# @TODO: Import libraries
import csv
import json
import os
from pathlib import Path
%matplotlib 
# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('menu_data.csv')
sales_filepath = Path('sales_data.csv')

menu = []
sales = []
report = {}

with open(menu_filepath, "r") as menulist:

    menufile_reader = csv.reader(menulist, delimiter = ',')
    menufile_header = next(menufile_reader)


    for row in menufile_reader:
        menu.append(row)

with open(sales_filepath, "r") as salelist:

    salefile_reader = csv.reader(salelist, delimiter = ',')
    salefile_header = next(salefile_reader)
    
    for row in salefile_reader:
        sales.append(row)
for row in sales:
    quantity = int(row[3])
    menuitem = row[4]

    if not menuitem in report:
        report[menuitem] = {         
        "01-count": 0,
        "02-revenue": 0,
        "03-cogs": 0,
        "04-profit": 0,
        }

    for mrow in menu:
        item = mrow[0]
        price = float(mrow[3])
        cost = int(mrow[4])
        profit = price - cost
        if menuitem == item:
            report[menuitem]["01-count"]+= quantity
            report[menuitem]["02-revenue"]+= price * quantity
            report[menuitem]["03-cogs"]+= cost *quantity
            report[menuitem]["04-profit"]+= profit * quantity

formatted = json.dumps(report, indent=4)
print(formatted)
#print(report)
output_path = 'output.txt'

# Open the output path as a file object
with open(output_path, 'w') as file:
     for key, value in report.items():  
        file.write('%s:%s\n' % (key, value))
```
